```python
# train_classifier.py
"""
Train helmet/seatbelt classifiers on Kaggle helmet data.
"""

import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from PIL import Image

# Config
DATA_DIR = "data/raw/helmet-detection"
CLASSIFIER_IMG_SIZE = (96, 96)
BATCH_SIZE = 16
EPOCHS = 5

# Dummy data loader – replace with actual loading from your Kaggle dataset
def load_helmet_data():
    # Return numpy arrays: X (images), y (0=no-helmet, 1=helmet)
    # This should loop through data/raw/helmet-detection/...
    # For now, create dummy data
    X = np.random.rand(100, 96, 96, 3).astype(np.float32)
    y = np.random.randint(0, 2, 100).astype(np.float32)
    return X, y

# Build MobileNetV3 classifier
base = keras.applications.MobileNetV3Small(
    input_shape=(96, 96, 3),
    include_top=False,
    weights='imagenet'
)
x = keras.layers.GlobalAveragePooling2D()(base.output)
x = keras.layers.Dense(128, activation='relu')(x)
output = keras.layers.Dense(1, activation='sigmoid')(x)
model = keras.Model(base.input, output)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

X_train, y_train = load_helmet_data()
model.fit(X_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE)

model.save("models/helmet_classifier.h5")
print("Helmet classifier saved to models/helmet_classifier.h5")
```
